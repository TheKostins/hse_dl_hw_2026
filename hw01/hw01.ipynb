```python
from torchvision.transforms import v2 
from IPython.display import Image
from torchvision import io
from torchvision.transforms.v2.functional import to_pil_image
```

```python
img = io.decode_image("./data/image.png")
to_pil_image(img)
```

```python
img.shape
```

```python
transforms = v2.Compose([
    v2.GaussianNoise(),
    v2.GaussianBlur(kernel_size=3),
    v2.RandomAffine(degrees=11),
    v2.RandomCrop(size=[150, 150], padding=[-10, -10]),
    v2.RandomSolarize(threshold=234)
])

transformed_img = transforms(img)
to_pil_image(transformed_img)
```

Вывод: с помощью аугментаций можно действительно искусственно увеличить размер датасета и улучшить качество модели.
